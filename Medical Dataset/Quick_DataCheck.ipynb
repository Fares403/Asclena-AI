## Read Data from csv files

```python
import pandas as pd 
import numpy as np

diagnosis = pd.read_csv('diagnosis.csv')
triage = pd.read_csv('triage_fixed.csv')
vitalsing = pd.read_csv('vitalsign_fixed.csv')
pyxis = pd.read_csv('pyxis.csv')
medrecon= pd.read_csv('medrecon.csv')
edstays = pd.read_csv('edstays.csv')
```

## list nulls for each table


```python
#list nulls for each table
tables = [diagnosis, pyxis, medrecon, edstays, triage, vitalsing]

for df in tables:
    print(df.isnull().sum())
    print('-----------------------------')
```

## list duplicates for each table

```python
# check for duplicates in each table
for df in tables:
    print(df.duplicated().sum())
    print('-----------------------------')
```
